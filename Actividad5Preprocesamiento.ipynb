7. ¿Hay valores atípicos de la variable `Weight`? Confírmalo con un gráfico de boxplot. ¿Cuántos son? Almacénalos en un dataframe independiente y elimínalos del conjunto original.

```python
response_df = pd.read_csv('responses.csv')
# grafico de valores atípicos con un boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=response_df['Weight'])
plt.title('Boxplot de Weight para identificar valores atípicos')
plt.xlabel('Weight')
plt.show()

# Identifica y almacena valores atípicos en otro DataFrame
Q1 = response_df['Weight'].quantile(0.25)
Q3 = response_df['Weight'].quantile(0.75)
IQR = Q3 - Q1
outliers_df = response_df[(response_df['Weight'] < Q1 - 1.5 * IQR) | (response_df['Weight'] > Q3 + 1.5 * IQR)]
num_outliers = len(outliers_df)
print(f"Número de valores atípicos en Weight: {num_outliers}")

# Eliminar valores atípicos del conjunto original
response_df.drop(outliers_df.index, inplace=True)

# grafico de valores atípicos con un boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=response_df['Weight'])
plt.title('Boxplot de Weight para identificar valores atípicos')
plt.xlabel('Weight')
plt.show()
```
